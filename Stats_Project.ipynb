# Player aging curves for elite NFL athletes in different positions


#Python Scripts for Histograms

Loading libraries

```python
%pip install -qq datascience
from datascience import *

import numpy as np
import matplotlib.pyplot as plots

%pip install -qq plotly
import plotly.graph_objects as go
import plotly.express as px

%matplotlib inline
plots.style.use(["seaborn-v0_8", "seaborn-v0_8-talk"])
```

## Loading Data for different Positions

### QB

```python
qb_url = "https://docs.google.com/spreadsheets/d/1qCi8OIc2brHv0wVwpAAwNbxQSJmE1n6vkKKPOnpiVvo/export?format=csv"
qb_data = Table.read_table(qb_url)
qb_data.show(5)
```

```python
player_ages_qb = qb_data.select("Name", "Age")
player_ages_qb.show(5)
```

```python
# Group by player and collect all their ages into lists
def collect_ages(ages):
    return list(ages)

player_age_lists_qb = player_ages_qb.group('Name', collect_ages)
player_age_lists_qb.show(5)
```

```python
player_age_lists_qb = player_age_lists_qb.with_column('Average Age',
                                                  player_age_lists_qb.apply(np.mean, 'Age collect_ages'))

player_age_lists_qb.show(5)
```

```python
qb_average_age = np.mean(player_age_lists_qb.column("Average Age"))
qb_average_age
```

```python
player_ages_qb.hist('Age', bins=np.arange(player_ages_qb.column('Age').min(),
                                        player_ages_qb.column('Age').max(), 1))
plots.xlabel('Age')
plots.ylabel('Number of Appearances')
plots.title('Distribution of Player Ages: QB')
plots.show()

age_counts = player_ages_qb.group('Age')
```

```python
# Box plot
plots.boxplot(player_ages_qb.column('Age'))
plots.ylabel('Age')
plots.xlabel('QB')
plots.title('Boxplot of Player Ages')
plots.show()
```
